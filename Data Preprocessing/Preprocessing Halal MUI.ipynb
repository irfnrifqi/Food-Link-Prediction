```python
# import packages
import numpy as np
import pandas as pd
```

```python
# dataframe hasil scraping
df = pd.read_json('Hasil Crawl Halal MUI.json')
df
```

```python
# membagi kolom produk ke beberapa kolom dengan "|" sebagai pemisah
new = df["nama_produk"].str.split("|", n = 6, expand = True)
```

```python
# membuat kolom "no"
df["no"]= new[0]
df.head()
```

```python
# membuat kolom "id"
df["id"]= new[1]
df.head()
```

```python
# membuat kolom "nama"
df["nama"]= new[2]
df.head()
```

```python
# membuat kolom produsen
df["produsen"]= new[3]
df.head()
```

```python
# membuat kolom sertif
df["sertif"]= new[4]
df.head()
```

```python
# membuat kolom tanggal
df["tanggal"]= new[5]
df.head()
```

```python
# membuat dataframe baru dengan hanya mengambil kolom "id", "nama", "produsen", dan "sertif"
df_new = df[['id', 'nama', 'produsen', 'sertif']]
df_new
```

```python
# export ke bentuk csv
df_new.to_csv(r'Hasil Process Halal MUI.csv',index=False)
```
