```python
#import package
import pandas as pd
import numpy as np
```

```python
# dataframe hasil scraping (without duplicates)
data = pd.read_json('Hasil Crawl KlikIndomaret.json', encoding='latin1')
data
```

```python
# dataframae 1 (untuk di merge dengan data tanpa takaran)
df1 = pd.read_csv('KlikIndomaret Manual Processed 1.csv', sep=';')
df1
```

```python
# dataframe 2 (produk tanpa takaran)
df2 = pd.read_csv('KlikIndomaret Manual Processed 2.csv', sep=';')
df2
```

```python
# merge antara 2 dataframe yang telah dibuat
joinmatch = pd.merge(df1, df2, left_on=df1['index'], right_on=df2['index'])
joinmatch
```

```python
# hapus kolom yang tidak dibutuhkan
joinmatch = joinmatch.drop(columns=['key_0', 'index_x', 'nama_produk_x', 'index_y', 'nama_produk_y'])
joinmatch
```

```python
# ganti nama Column3 menjadi produk
joinmatch.rename(columns={"Column3": "produk"}, inplace=True)
joinmatch
```

```python
# export ke bentuk csv
joinmatch.to_csv(r'Data KlikIndomaret Ready.csv')
```
